```python
from wordcloud import WordCloud
import pandas as pd

# Display the generated image:
# the matplotlib way:
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pipeline/data/all_the_news/articles2.csv')
df.head()
```

```python
text = df.loc[55].content[:1000]
```

```python
# Generate a word cloud image
wordcloud = WordCloud(background_color=None, mode='RGBA', colormap='seismic').generate(text)
plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# lower max_font_size
wordcloud = WordCloud(max_font_size=40, background_color=None, mode='RGBA', colormap='viridis').generate(text)
plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
